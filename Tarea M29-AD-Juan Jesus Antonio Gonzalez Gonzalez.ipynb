```python
#se tiene un ejemplo ademas de usar request se usa la libreria Beatifulsoup, que permite
#una impresion mas amigable de contenidos de este tipo

import requests
from bs4 import BeautifulSoup
from pprint import pprint

#esto se usa para simular que somos humanos, y para que el sitio no se de cuenta que lo estamos scrapping
#no siempre se debe utilizar pero es recomendado

headers= {
    'Access-Control-Allow-Origin':'*',
    'Access-Control-Allow-Methods':'GET',
    'Access-Control-Allow-Headers':'Content-Type',
    'Access-Control-Max-Age':'3600',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (HTML, like Gecko) Chrome/131.0.0.0 Safari/537.36'

}
url='https://www.animenewsnetwork.com/encyclopedia/api.php'
data=requests.get(url,headers)

#variable para guardar los articulos del sitio web
url_data=[]

html=BeautifulSoup(data.text,'html.parser')
articles = html.select('a.post-cart')

for article in articles:
    title = article.select('.card-title')[0].get_text()
    excerpt = article.select('.card-text')[0].get_text()
    pub_date=article.select('.card-footer small')[0].get_text()

    url_data.append({"title": title, "excerpt": excerpt,"pub_date": pub_date})

pprint(url_data)
```

```python
import requests
from bs4 import BeautifulSoup

# URL de la p√°gina web
url = "https://quotes.toscrape.com/"

# Simular que somos un navegador para evitar bloqueos
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}

# Realizar la solicitud GET
response = requests.get(url, headers=headers)

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    # Parsear el contenido HTML con BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Extraer los t√≠tulos de las citas (etiquetas <span> con clase "text")
    quotes = soup.find_all('span', class_='text')
    
    # Imprimir las citas encontradas
    for i, quote in enumerate(quotes, start=1):
        print(f"Cita {i}: {quote.get_text()}")
else:
    print(f"Error: {response.status_code}")

```

```python
from bs4 import BeautifulSoup
import requests

print("\n### Ejercicio 2: Parseo de informaci√≥n con BeautifulSoup")
url_mercadolibre = "https://listado.mercadolibre.com.mx/laptops"
response = requests.get(url_mercadolibre)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    items = soup.find_all('h2', class_='ui-search-item__title')  # Ejemplo de parseo
    
    print("Listado de items encontrados:")
    for i, item in enumerate(items[:10]):  # Limitar a los primeros 10 resultados
        print(f"{i+1}. {item.text}")
else:
    print(f"Error al acceder a MercadoLibre: {response.status_code}")
```

```python
#Generalmente las respuestas de una llamada API son archivos J50N 
 
import json 
#Se genera una funci√≥n simple llamada "jprint" 
def jprint(obj):
    # create a formatted string of the Python JSON object 
    text = json.dumps(obj, sort_keys=True, indent=4) 
    print(text) 
jprint(response_API.json())
```
