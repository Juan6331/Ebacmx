````python
#Revisar el EDA del ejercicio anterior y añadir siguientes elementos, basados en los últimos 4 años

#Crecimiento:

#1) qué género ha tenido el mayor y menor crecimiento,
#2) qué plataforma ha tenido el mayor y menor crecimiento,
#3) qué videojuego ha tenido el mayor y menor crecimiento
#2 elementos adicionales que muestren oportunidades de mercado

# Filtrar últimos 4 años disponibles en el dataset
latest_year = df["Year"].max()
df_recent = df[df["Year"] >= (latest_year - 3)]  # Últimos 4 año
````

````python
# Agrupar ventas de los últimos 4 años por Género
genre_growth = df_recent.groupby("Genre")["Global_Sales"].sum().reset_index()

# Ordenar por ventas y calcular crecimiento
genre_growth["Crecimiento (%)"] = genre_growth["Global_Sales"].pct_change() * 100

# Identificar mayor y menor crecimiento
genre_max = genre_growth.loc[genre_growth["Crecimiento (%)"].idxmax()]
genre_min = genre_growth.loc[genre_growth["Crecimiento (%)"].idxmin()]

print(f"📈 Mayor crecimiento: {genre_max['Genre']} ({genre_max['Crecimiento (%)']:.2f}%)")
print(f"📉 Menor crecimiento: {genre_min['Genre']} ({genre_min['Crecimiento (%)']:.2f}%)")
````

The active selection performs a series of operations to analyze the growth in global sales of video games by genre over the last four years.

1. **Grouping Sales by Genre**:
   ```python
   genre_growth = df_recent.groupby("Genre")["Global_Sales"].sum().reset_index()
   ```
   This line groups the data in the DataFrame `df_recent` by the "Genre" column and sums the "Global_Sales" for each genre. The `groupby` method is used to group the data, and the `sum` method aggregates the sales. The `reset_index` method is called to convert the grouped data back into a DataFrame with a default integer index.

2. **Calculating Growth Percentage**:
   ```python
   genre_growth["Crecimiento (%)"] = genre_growth["Global_Sales"].pct_change() * 100
   ```
   This line calculates the percentage change in global sales for each genre and adds it as a new column "Crecimiento (%)" in the `genre_growth` DataFrame. The `pct_change` method computes the percentage change between the current and prior element, and multiplying by 100 converts it to a percentage.

3. **Identifying Maximum and Minimum Growth**:
   ```python
   genre_max = genre_growth.loc[genre_growth["Crecimiento (%)"].idxmax()]
   genre_min = genre_growth.loc[genre_growth["Crecimiento (%)"].idxmin()]
   ```
   These lines identify the genres with the highest and lowest growth percentages. The `idxmax` method returns the index of the first occurrence of the maximum value in the "Crecimiento (%)" column, and `idxmin` returns the index of the first occurrence of the minimum value. The `loc` method is then used to select the rows corresponding to these indices.

4. **Printing Results**:
   ```python
   print(f"📈 Mayor crecimiento: {genre_max['Genre']} ({genre_max['Crecimiento (%)']:.2f}%)")
   print(f"📉 Menor crecimiento: {genre_min['Genre']} ({genre_min['Crecimiento (%)']:.2f}%)")
   ```
   Finally, these lines print the genres with the highest and lowest growth percentages, formatted to two decimal places. The `f-string` formatting is used to embed expressions inside string literals, making the output more readable and informative.

Overall, this code provides a clear analysis of which video game genres have experienced the most significant growth and decline in global sales over the specified period.
